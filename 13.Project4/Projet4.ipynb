```python
# --- Gráfico 1: Distribuição da variável alvo 'Churn' ---

# Usando o countplot para contar e plotar as ocorrências de cada categoria
# O parâmetro 'order' organiza as barras da mais frequente para a menos frequente
sns.countplot(x='Churn', data=df, order=df['Churn'].value_counts().index)
# Adicionando título e rótulos aos eixos
plt.title('Distribuição de Churn (Cancelamentos)')
plt.xlabel('O Cliente Cancelou?')
plt.ylabel('Contagem')
```

```python
# --- Gráfico 3: Distribuição do Serviço de Internet ---

sns.countplot(x='Servico_Internet', data=df, order=df['Servico_Internet'].value_counts().index)
plt.title('Distribuição por Serviço de Internet')
plt.xlabel('Serviço de Internet')
plt.ylabel('Contagem')
```

```python
# --- Gráfico 4: Distribuição do Suporte Técnico ---

sns.countplot(x='Suporte_Tecnico', data=df, order=df['Suporte_Tecnico'].value_counts().index)
plt.title('Distribuição por Suporte Técnico')
plt.xlabel('Possui Suporte Técnico?')
plt.ylabel('Contagem')
```

```python
# Criando uma lista com os nomes das variáveis que queremos analisar
boolean_vars = ['Genero', 'Idoso', 'Casado', 'Dependentes', 'Churn']

# --- Parte 1: Imprimir a contagem de valores no console ---
# (Esta parte gera o texto com os números que mostrei antes dos gráficos)
print("Contagem de valores para as variáveis Booleanas:")
# Um loop 'for' para passar por cada variável da nossa lista
for var in boolean_vars:
    # Um tratamento especial para a coluna 'Churn', para contar os valores faltantes
    if var == 'Churn':
        counts = df[var].value_counts(dropna=False)
    else:
        counts = df[var].value_counts()
    print(f"\n----- {var} -----")
    print(counts)

# --- Parte 2: Criar os gráficos de pizza ---
# Configurando o tamanho da figura que vai conter todos os gráficos
plt.figure(figsize=(20, 8))
# Adicionando um título geral para a figura
plt.suptitle('Balanceamento das Variáveis Booleanas', fontsize=20)

# Um loop 'for' para criar um gráfico para cada variável
# 'enumerate' nos dá um contador (i) e o nome da variável (var) a cada iteração
for i, var in enumerate(boolean_vars):
    # Criando um subplot em uma grade de 2 linhas e 3 colunas. A posição (i+1) muda a cada loop.
    plt.subplot(2, 3, i + 1)
    
    # Pegando os dados de contagem novamente para cada variável
    if var == 'Churn':
        # 'dropna=False' inclui os valores nulos na contagem
        counts = df[var].value_counts(dropna=False).rename(index={pd.NA: 'Faltante'})
    else:
        counts = df[var].value_counts()
        
    # A função que desenha o gráfico de pizza
    plt.pie(counts,                               # Os dados (quantidades de cada categoria)
            labels=counts.index,                  # Os rótulos de cada fatia
            autopct='%1.1f%%',                    # O formato para mostrar a porcentagem
            startangle=90,                        # O ângulo inicial para a primeira fatia
            colors=['skyblue', 'lightcoral', 'lightgrey']) # Cores personalizadas
            
    # Adicionando um título para cada subplot individual
    plt.title(f'Distribuição de {var}', fontsize=14)
    # Removendo o texto 'ylabel' que o matplotlib adiciona por padrão
    plt.ylabel('')

# Ajustando o layout para evitar sobreposição de texto
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
# --- Metodo de pagamento x Churn ---

plt.figure(figsize=(10, 7))
sns.countplot(y='Metodo_de_Pagamento', hue='Churn', data=df)
plt.title('Relação entre Método de Pagamento e Churn', fontsize=14)
plt.xlabel('Contagem de Clientes')
plt.ylabel('Método de Pagamento')
plt.legend(title='Churn')
plt.tight_layout()
```

Nessa comparação de formas de pagamentos, claramente é possivel verificar que Eletronic check "Boleto ou cheque digital" e que nos trás maior Churn dos cliente, e o que mais reten o cliente são as formas de pagamento automaticos por trazer uma comodidade de pagamentos. 

```python
# --- StreamingTV x Churn ---

plt.figure(figsize=(10, 6))
sns.countplot(x='StreamingTV', hue='Churn', data=df)
plt.title('Relação entre Serviço de Streaming de TV e Churn', fontsize=14)
plt.xlabel('Possui Streaming de TV?')
plt.ylabel('Contagem de Clientes')
plt.legend(title='Churn')
plt.tight_layout()
```

A análise dos gráficos revela que os resultados são bastante similares, com uma pequena diferença: a taxa de churn é ligeiramente maior entre os que não contrataram streaming. Destaca-se que a melhor taxa de conversão pertence aos que contrataram apenas o streaming de TV.

```python
# --- Tempo de cliente  x Churn ---

# Criando os grupos (bins)
bins = [-1, 12, 36, 60, 100]
# -1 usado para garantir que clientes com 0 mese seja incluso
labels = ['Novo (0-12m)', 'Intermediário (13-36m)', 'Leal (37-60m)', 'Fiel (>60m)']
df['Tempo_Cliente_Bin'] = pd.cut(df['Tempo_como_Cliente'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.countplot(x='Tempo_Cliente_Bin', hue='Churn', data=df.sort_values('Tempo_Cliente_Bin'))
plt.title('Relação entre Tempo como Cliente e Churn', fontsize=14)
plt.xlabel('Nível de Antiguidade do Cliente')
plt.ylabel('Contagem de Clientes')
plt.legend(title='Churn')
plt.tight_layout()
plt.savefig('pergunta_4_tempo_cliente_churn.png')
```

A análise indica que a probabilidade de cancelamento diminui conforme aumenta o tempo de permanência do cliente. Isso evidencia a necessidade de um pós-venda eficiente para garantir a retenção, especialmente nos primeiros meses.

```python
# --- Tempo de cliente  x Churn ---

g = sns.catplot(x='Tipo_Contrato', hue='Suporte_Tecnico', col='Churn',
                data=df, kind='count',
                height=6, aspect=1,
                legend=False)
g.fig.suptitle('Relação Contrato x Suporte Técnico para cada Status de Churn', y=1.03, fontsize=14)
g.set_axis_labels('Tipo de Contrato', 'Contagem de Clientes')
g.set_titles("Clientes com Churn = {col_name}")
g.add_legend(title='Possui Suporte Técnico')
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig('pergunta_5_contrato_suporte_churn.png')
```

Clientes com contrato mensal e sem suporte técnico representam o maior risco de cancelamento. Em contraste, a maioria dos clientes que permanecem possui contratos de um ou dois anos, demonstrando que vínculos mais longos e com apoio ao cliente são cruciais para a retenção.
